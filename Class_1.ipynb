### Create a Data frame

```python
# Crreate a Data frame
df = pd.DataFrame()
```

```python
# Adding the columns
df['Name'] = ['Nivi','Ana','Isabella','Mark']
df['Sex'] = ['Male','Female','Female','Male']
df['Country'] = ['India','USA','Spain','UK']
df['Color'] = ['Blue','Pink','Orange','Red']
df['Number'] = [7,5,3,11]
```

```python
df
```

```python
type(df)
```

```python
# Create the new rows
new = pd.Series(['Angle','Female','Australia','Brown',1], index=['Name','Sex','Country','Color','Number'])
```

```python
# Add the previous series using append()
df.append(new,ignore_index=True)
```
